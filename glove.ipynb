```python
import spacy
import pandas as pd

nlp = spacy.load('en_core_web_lg') 
df = pd.read_csv("data/df_train.csv")  

print(df.head())

def count_words_with_glove_embeddings(tokens, nlp):
    count_with_glove = 0
    total_words = 0
    for token in tokens:
        total_words += 1  
        if nlp.vocab[token].has_vector:
            count_with_glove += 1
    return count_with_glove, total_words


df["tokens"] = df["tokens"].apply(eval) 
df[["words_with_glove", "total_words"]] = df["tokens"].apply(lambda tokens: pd.Series(count_words_with_glove_embeddings(tokens, nlp)))
total_words_with_glove = df["words_with_glove"].sum()
total_words_in_dataset = df["total_words"].sum()

# 7. Calcular el rango
print(f"Total de palabras con embeddings en GloVe: {total_words_with_glove}")
print(f"Total de palabras en el dataset: {total_words_in_dataset}")
print(f"Rango de palabras con embeddings: {total_words_with_glove}/{total_words_in_dataset}")
print(f"Porcentaje de palabras con embeddings en GloVe: {(total_words_with_glove / total_words_in_dataset) * 100:.2f}%")
```
